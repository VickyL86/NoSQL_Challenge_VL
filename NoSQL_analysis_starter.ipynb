```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
#pretty print the first document in the collection
establishments.find_one()
```

```python
#save establishments collection in a csv file
df = pd.DataFrame(list(establishments.find()))
df.to_csv('establishments.csv', index=False)
```

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
print("Number of establishments with a hygiene score of 20:",establishments.count_documents(query))

# Display the first document in the results using pprint
print("First Result:")
results = establishments.find(query)
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
Hygiene_20 = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print("Number of rows in DataFrame:",len(Hygiene_20))
# Display the first 10 rows of the DataFrame
Hygiene_20.head(10)
```

```python
#export dataframe to csv    
Hygiene_20.to_csv('Hygiene_20.csv', index=False)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
#double check by print number of records where RatingValue is an float
print("Number of records where RatingValue is a float:",establishments.count_documents({'RatingValue': {'$type': 'double'}}))
```

```python
# Find the establishments with 'City of London Corporation' as the Local Authority and has a RatingValue greater than or equal to 4.
query_rating = {'$and': [{'LocalAuthorityName': 'City of London Corporation'}, {'RatingValue': {'$gte': 4}}]}
# Use count_documents to display the number of documents in the result
print("Number of establishments in London with rating greater than or equal to 4:",establishments.count_documents(query_rating))
print("First Result:")
results_rating = establishments.find(query_rating)
pprint(results_rating[0])
```

```python
# Convert the result to a Pandas DataFrame
London_Rating = pd.DataFrame(results_rating)

# Display the number of rows in the DataFrame
print("Number of rows in DataFrame:",len(London_Rating))

# Display the first 10 rows of the DataFrame
London_Rating.head(10)
```
